```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Absolute path of the CSV file
file_path = r'C:\Users\rsald\Dropbox\Study Courses\Udemy\Power BI Bootcamp - Build Real World Power BI Projects\Project 6 - Bank Data Analysis\P6-UK-Bank-Customers.csv'

# Column names
gender_column = 'Gender'
balance_column = 'Balance'
region_column = 'Region'
job_classification_column = 'Job Classification'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Perform analysis on the data
# 1. Examine the distribution of account balances between male and female customers
balance_by_gender = df.groupby(gender_column)[balance_column].mean()
print(balance_by_gender)

# Visualize balance distribution by gender
sns.barplot(x=balance_by_gender.index, y=balance_by_gender.values)
plt.title('Average Account Balance by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Account Balance')
plt.show()

# 2. Identify regions with the highest number of bank accounts
accounts_by_region = df[region_column].value_counts()

# Visualize accounts by region
accounts_by_region.plot(kind='bar')
plt.title('Number of Bank Accounts by Region')
plt.xlabel('Region')
plt.ylabel('Number of Bank Accounts')
plt.xticks(rotation=45)
plt.show()

# 3. Explore the distribution of account holders across different job classifications
accounts_by_job_classification = df[job_classification_column].value_counts()

# Visualize accounts by job classification
plt.figure(figsize=(8, 6))
accounts_by_job_classification.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Account Holders by Job Classification')
plt.ylabel('')
plt.show()
```
